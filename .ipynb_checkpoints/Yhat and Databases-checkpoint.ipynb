## Training a model

```r
library("randomForest")

# isolate target data
churn.result <- factor(df$Churn.)

# drop columns not relavent to the statistical model
to.drop <- c('State','Area.Code','Phone','Churn.')
churn.feat.space <- df[,!(names(df) %in% to.drop)]

# convert "yes"/"no" columns to TRUE/FALSE values
yes.no.cols <- c("Int.l.Plan","VMail.Plan")
churn.feat.space[yes.no.cols] <- churn.feat.space[yes.no.cols] == "yes"

# pull out column names for future use
feats <- names(churn.feat.space)

# train a random forest model
clf <- randomForest(churn.feat.space,y=churn.result)
```

As a caveat, I've pulled out the column names of the feature `data.frame` and stored them in the variable `feats`.

## Deploying to ŷhat

Finally I can declare my ŷhat model by defining `model.require`, `model.transform`, and `model.predict`.

```r
# which libraries should yhat require?
model.require <- function() {
    library(randomForest)
}

# how should yhat transform the incoming data.frame?
model.transform <- function(df) {
    df <- df[feats]
    df[yes.no.cols] <- df[yes.no.cols] == "yes"
    df 
}

# execute the prediction function
model.predict <- function(df) {
    churn.prob <- predict(clf,newdata=df,type="prob")
    churn.risk <- churn.prob[colnames(churn.prob) == "True."]
    customer.value <- rowSums(df[c("Day.Charge","Eve.Charge","Night.Charge")])
    expected.loss <- churn.risk * customer.value
    data.frame(churn.risk,customer.value,expected.loss)
}

# pass yhat module credentials and deploy to server
yhat.config  <- c(
    username="USERNAME",
    apikey="APIKEY",
    env="ENTERPRISE URL"
)
yhat.deploy("RChurnModel")
```

## Predicting on a query

I'll be using Python to cordinate dataflow between my MySQL database and deployed ŷhat model. Specifically, the Pandas library, an open source package geared towards tabular data analysis. Because of it's many strengths we've designed ŷhat to interact well with (if not prefer) Pandas __Data Frames__.

Using the `MySQLdb` library I can create a connection to the database for `pandas` to use to execute and read in the result of a query. Critically, what's return by the `read_frame` function is a Pandas `DataFrame` instance.

````python
import MySQLdb      # `pip install MySQL-python`
import pandas as pd # `pip install pandas`
con = MySQLdb.connect(user="e", passwd="blah", db="churn", host="localhost")

# Query database and return a pandas' DataFrame
prediction_query = "SELECT * FROM userdata LIMIT 5"
df = pd.io.sql.read_frame(prediction_query,con)
df
````

Though ŷhat models ingest JSON through REST calls, we've designed both the Python and R clients to abstract away from the actual network request. From either Python or R you can simply pass, respectively, a `pandas.DataFrame` or `data.frame` object to the client's predict function.

````python
from yhat import Yhat # `pip install yhat`
yh = Yhat(
    "eric@yhathq.com",
    "328e2649f535c8ba992e413a5a307c96",
    "http://sandbox.yhathq.com/"
)
# Pass DataFrame to model
response = yh.predict("RChurnModel",df)

# "pretty" print using json.dumps()
import json
print json.dumps(response,indent=2)
````

## Writing results back to a database

Finally, we'll take the model predictions and write them back out to a database table.



````python
result_df = pd.DataFrame(response["result"])
result_df
````

````python
pd.io.sql.write_frame(result_df,"churn_result",con,flavor="mysql",if_exists="replace")
````
