# Use Cases: Nightly Redeployment

I'm the owner of an ecommerce website and would like to use ŷhat to implement an item recommender. To help with this, I've allowed users to "star" items they're interested in on my website, allowing me to track and compare users' preferences.

Each time a user stars an item, those results are stored on database (for this example we'll say MySQL), a process which might look something like this:

<img src="files/user_star_table.png" width=600>

While it may be easy to build and deploy a recommender once, a static recommender model often becomes outdated quickly. Consequently,  I'd like my recommender to be updated as new items are added to the shop and users produce more data. To achieve this, my recommender model must be retrained periodically and then redeployed to keep the recommendations current.

## Training and deploying a recommender

My ŷhat model will train by pulling directly from my MySQL database table of user/item relationships. It will then compute item similarities and finally deploy a model to the ŷhat server for my website to use. Passing the list of items starred by a user will produce a list of recommended items that a specific user may also like.

We won't be going in depth about the following code because the primary concepts are covered extensively in other documentation. You can find more information on how ŷhat interacts with databases [here](http://nbviewer.ipython.org/github/yhat/use-cases/blob/master/yhat-and-dbs/yhat-and-databases.ipynb), along with an in depth discussion of recommenders [here](http://blog.yhathq.com/posts/recommender-system-in-r.html).

---
> FILENAME: `/home/eric/retrain_model.py`

----
```python
#!/usr/bin/env python

# The header above is important since it lets bash know what command
# to call (python) in order to run this file 

import MySQLdb
import pandas as pd
import numpy as np

from sklearn.metrics.pairwise import cosine_similarity

con = MySQLdb.connect(user="e", passwd="password", 
                      db="stars", host="localhost")

query = "SELECT user_id, item_id FROM userstarstable"

# read from database
user_stars = pd.io.sql.read_frame(query,con)
user_stars["value"] = np.ones(len(user_stars))

# construct pivot table
stars_table = pd.pivot_table(user_stars, values = ["value"],
                             rows = ["item_id","user_id"],
                             aggfunc = np.max).unstack()
stars_table = stars_table.fillna(0)

# calculate item similarities
item_sims = cosine_similarity(stars_table)
item_names = stars_table.index.values
item_sims = pd.DataFrame(item_sims, columns = item_names, index = item_names)

# helper function to get similiar items given a list of starred items
def get_similar_items(starred_items):
    # calculate item similarities
    rec_items = item_sims[starred_items].mean(axis=1).order(ascending=False).index
    # filter out items passed
    is_new_item = rec_items.map(lambda item: item not in starred_items)
    return rec_items[is_new_item]

from yhat import Yhat,YhatModel,preprocess

# Define a YHat model
class ItemRecommender(YhatModel):
    @preprocess(in_type=dict,out_type=dict)
    def execute(self,data):
        starred_items = data["starred_items"]
        similar_items = get_similar_items(starred_items)
        return {"recommended_items":similar_items}

# Deploy YHat model to server
yh = Yhat("USERNAME","APIKEY","ENTERPRISE URI")
yh.deploy("ItemRecommender",ItemRecommender,globals())

```

## Nightly builds with cron

There are many ways to perform recurring jobs, but the Linux/Unix cron scheduler is by far the simplest. With cron, commands can easily be created to run periodically. In this particular case I'd like to retrain my recommender nightly to keep my recommendations up-to-date with my product offerings.

Let's say I've chosen to retrain my recommender nightly at 4:00a.m. Within my Linux file system, I'll construct a cron table with a single job. For this example I've named my cron table "`yhat.crontab`" while the script to construct a ŷhat model sits in "`/home/eric/retrain_model.py`".

---
> FILENAME: `yhat.crontab`

---

```bash
# m h dom mon dow command
00 04  * * *  /home/eric/retrain_model.py >> /home/eric/retrain_model.log 2>&1
```

---
> ENDFILE

---

The first five rows of the file determine how often the trailing command should run. A numeric value will bind the job to some specific instance of a cycle (e.g. the 4th hour of a day), while a wildcard (`*`) matches any time period. The particular job above activates at 4:00am on every day of the year.

Furthermore, I've redirected the STDOUT and STDERR of the python script to a log file for debugging purposes. This will ensure I have a place to find the output in the case of an error.

To initialize the cron job, I first make my python script executable and then use the `crontab` command to assign the cron table to a user (in this case "eric"). Once I've executed those commands, the job is now scheduled and users can begin seeing recommendations!

```bash
$ chmod +x /home/eric/retrain_model.py
$ crontab -u eric yhat.crontab
```

To stop the cron job from running execute the following command. Be careful though, this will remove all cron jobs tied to that user.


```bash
$ crontab -u eric -r
```


## How ŷhat transitions to new model versions

Though this model trains late in the evening, there's no down time when deploying new versions. As new versions of my recommender are deployed, ŷhat enterprise will wait until the model is fully built before transitioning to the new version. This means that users on your webpage during new model deployments won't be interrupted and will still get product recommendations.
