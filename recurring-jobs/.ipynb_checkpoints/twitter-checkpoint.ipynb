```python
import tweepy
import json
import pandas as pd
import numpy as np
import time
```

```python
creds = json.load(open("creds.json",'r'))
auth = tweepy.OAuthHandler(creds["consumer_key"],
                           creds["consumer_secret"])
auth.set_access_token(creds["access_token"],
                      creds["access_token_secret"])
api = tweepy.API(auth)

public_tweets = api.home_timeline()
for i,tweet in enumerate(public_tweets):
    print "[%d] %s" % (i,tweet.text)
```

```python
def tweet_to_dict(tweet):
    author = tweet.author
    return {
        "tweet_id":tweet.id,
        "username":author.name,
        "text":tweet.text,
        "verified":author.verified,
        "followers":author.followers_count
    }
tweet_to_dict(api.search("#neighbors")[0])
```

```python
df = pd.DataFrame(map(tweet_to_dict,api.search("#neighbors",count=100)))
min_id = df.tweet_id.min()
df = df.append(pd.DataFrame(map(tweet_to_dict,api.search("#neighbors",count=100,max_id=min_id))))
```

```python
def clean_tweets(df,text_fields):
    df[text_fields] = df[text_fields].applymap(lambda t:t.encode('utf-8'))
    return df
```

```python
n_iter = 10
min_tweets = 1000
query = "@yhathq"
df = None
while(True):
    if df is None:
        df = pd.DataFrame(map(tweet_to_dict,api.search(query,count=100)))
    else:
        df = df.append(pd.DataFrame(map(tweet_to_dict,api.search(query,count=100,max_id=min_id))))
    min_id = df.tweet_id.min()
    if len(df) > min_tweets:
        break
df = clean_tweets(df,["text","username"])
df['sentiment'] = np.zeros(len(df))
```
