# Use Cases: ŷhat and Databases

Though ŷhat can handle arbitrary data structures, we've taken deliberate steps to integrate well with tabular data. This example use case will attempt to demonstrate how ŷhat can be used to systematically bridge the gap between predictive models and relational databases. The example specifically covered is a simple read, predict, write pipeline where the results of a custom query is fed to a model then written back out to another table.

## Training a model

There are a lot of steps between gathering data and having a working model to deploy, but for the sake of brevity this article will skip over that process. For now, let's assume that an analyst has done validation, and wishes to push his/her model to production.

Using R, the following code cleans the `data.frame` and trains a random forest classifier. This classifier will be the main engine behind the predictive model I'll be deploying to my ŷhat server.

```r
library("randomForest")

# isolate target data
churn.result <- factor(df$Churn.)

# drop columns not relavent to the statistical model
to.drop <- c('State','Area.Code','Phone','Churn.')
churn.feat.space <- df[,!(names(df) %in% to.drop)]

# convert "yes"/"no" columns to TRUE/FALSE values
yes.no.cols <- c("Int.l.Plan","VMail.Plan")
churn.feat.space[yes.no.cols] <- churn.feat.space[yes.no.cols] == "yes"

# pull out column names for future use
feats <- names(churn.feat.space)

# train a random forest model
clf <- randomForest(churn.feat.space,y=churn.result)
```

As a caveat, I've pulled out the column names of the feature `data.frame` and stored them in the variable `feats`. This will be very useful when implementing my ŷhat model in case future columns are out of order, or extra columns have been added.

## Deploying to ŷhat

Finally I can declare a ŷhat model by defining `model.require`, `model.transform`, and `model.predict`. These replicate the work done in the previous section and add a call the random forest classifier. A `data.frame` is returned containing the churn risk of each row (probability of churn), the customer value, and expected loss. The R ŷhat client auto-detects variables used in these methods and captures them before being deployed to the server. This means that the `clf` variable (a.k.a. the trained random forest) will be captured during the `yhat.deploy()` call.

```r
# which libraries should yhat require?
model.require <- function() {
    library(randomForest)
}

# how should yhat transform the incoming data.frame?
model.transform <- function(df) {
    df <- df[feats] # Filter features by column
    df[yes.no.cols] <- df[yes.no.cols] == "yes"
    df 
}

# execute the prediction function
model.predict <- function(df) {
    churn.prob <- predict(clf,newdata=df,type="prob")
    churn.risk <- churn.prob[colnames(churn.prob) == "True."]
    customer.value <- rowSums(df[c("Day.Charge","Eve.Charge","Night.Charge")])
    expected.loss <- churn.risk * customer.value
    data.frame(churn.risk,customer.value,expected.loss)
}

# pass yhat module credentials and deploy to server
yhat.config  <- c(
    username="USERNAME",
    apikey="APIKEY",
    env="ENTERPRISE URL"
)
yhat.deploy("RChurnModel")
```

## Predicting on a query

Now that we have a deployed model it's time to predict on SQL data.

I'll be using Python to cordinate dataflow between my MySQL database and deployed ŷhat model. Specifically, the Pandas library, an open source package geared towards tabular data analysis. Because of it's many strengths we've designed ŷhat to interact well with (if not prefer) Pandas __Data Frames__.

Using the `MySQLdb` library I can create a connection to the database for `pandas` to use to execute and read in the result of a query. The query itself is simple, only reading five rows, but can be far more complex for more peculiar needs. Importantly, the `read_frame` function returns a Pandas `DataFrame` instance which is essentially synonymous with the tabular structure of the SQL database.

````python
import MySQLdb      # `pip install MySQL-python`
import pandas as pd # `pip install pandas`
con = MySQLdb.connect(user="e", passwd="password", db="churn", host="localhost")

# Query database and return a pandas' DataFrame
prediction_query = "SELECT * FROM userdata LIMIT 5"
query_result = pd.io.sql.read_frame(prediction_query,con)
query_result
````

Though ŷhat models ingest JSON through REST calls, we've designed both the Python and R clients to abstract away from the actual network request. From either Python or R you can simply pass, respectively, a `pandas.DataFrame` or `data.frame` object to the client's predict function.

In this case I'll pass the result of my MySQL query to the __RChurnModel__ and then print the JSON response.

````python
from yhat import Yhat # `pip install yhat`
yh = Yhat(
    "USERNAME",
    "APIKEY",
    "ENTERPRISE URL"
)
# Pass DataFrame to model
response = yh.predict("RChurnModel",query_result)

# "pretty" print using json.dumps()
import json
print json.dumps(response,indent=2)
````

## Writing results back to a database

Finally, we'll take the model predictions and write them back out to a database table.

A good first step would be to convert the JSON structure to a tabular DataFrame. This boils down to initializing a DataFrame object with the `"result"` field of the response.

````python
# Convert JSON result to a DataFrame
result_df = pd.DataFrame(response["result"])
result_df
````

Once I have a DataFrame, I can write it back to MySQL using Pandas and the connection object created earlier. For good measure, I've run a `%sql` select statement to ensure the results were written.

````python
pd.io.sql.write_frame(result_df,"churn_result",con,flavor="mysql",if_exists="replace")
%sql select * from churn_result
````

## Final Comments

Pulling and writing from databases is nothing novel. However, because the business logic of pulling from a database is abstracted from the code used to execute a prediction, models and query methodologies can easily change without the need to update the other.
